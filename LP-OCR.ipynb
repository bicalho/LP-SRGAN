# Run OCR on all license plates

Iterate through all the license plates in the dataset folder and run OCR in the image in 4 different scales, then compare them with the label in the file name

```python
import glob
import matplotlib.pyplot as plt
import cv2
import numpy as np

import pytesseract as pt

from PIL import Image
pt.pytesseract.tesseract_cmd = 'C:\Program Files\Tesseract-OCR\\tesseract.exe'
```

```python
data_gan = glob.glob('./gan_images/*')
data_bicubic = glob.glob('./bicubic_images/*')
data = glob.glob('./datasets/LP/*')
```

## Post proccess the ocr

```python
def pos_proc_ocr(ocr):
    ocr = ocr.replace('|','1')
    ocr = ocr.replace(' ','')
    ocr = ocr[:3].replace('1','I').replace('!','I') + ocr[3:]
    ocr = ocr[:3].replace('0','O') + ocr[3:]
    ocr = ocr[:3]+ ocr[3:].replace('O','0').replace('I','1').replace('!','1')
    
    ocr_final = ocr
    return ocr_final
```

## Run the ocr on the dataset files
- Get the ocr of the original file
- Reduce the image in 4,6,8 scales
- Store the reduced image
- Run the ocr on the new image

```python
import pandas as pd
zipped_list = list(zip(original_list,ocr_list, gan_list, bic_list))

ocr_df = pd.DataFrame(zipped_list, columns = ['Real', 'OCR 1x', 'OCR GAN', 'OCR BICUBIC'])
ocr_df.to_csv('ocr.csv', index = False, header = True)
```
