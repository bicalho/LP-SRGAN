# Run OCR on all license plates

Iterate through all the license plates in the dataset folder and run OCR in the image in 4 different scales, then compare them with the label in the file name

## Import libraries

```python
import glob
import matplotlib.pyplot as plt
import cv2
import numpy as np

import pytesseract as pt

from PIL import Image
```

```python
data = glob.glob('./generate_dataset/dataset/*')
```

## Post proccess the ocr

```python
def pos_proc_ocr(ocr):
    ocr = ocr.replace('|','1')
    ocr = ocr.replace(' ','')
    ocr = ocr[:3].replace('1','I').replace('!','I') + ocr[3:]
    ocr = ocr[:3].replace('0','O') + ocr[3:]
    ocr = ocr[:3]+ ocr[3:].replace('O','0').replace('I','1').replace('!','1')
    
    ocr_final = ocr
    return ocr_final
```

## Run the ocr on the dataset files
- Get the ocr of the original file
- Reduce the image in 4,6,8 scales
- Store the reduced image
- Run the ocr on the new image

```python
original_list = []
ocr_list = []
ocr_4_list = []
ocr_6_list = []
ocr_8_list = []

for path in data:
    try:
        plate_name = path.split('\\')
        save_name = plate_name[1]
        plate_name = plate_name[1].split('.')[0]
        img = cv2.imread(path)
        ocr_real = pt.image_to_string(img).upper()
        ocr_real = pos_proc_ocr(ocr_real)
    #     img_4x = cv2.resize(img, (int(img.shape[1]/4), int(img.shape[0]/4)), interpolation = cv2.INTER_AREA)
    #     img_6x = cv2.resize(img, (int(img.shape[1]/6), int(img.shape[0]/6)), interpolation = cv2.INTER_AREA)
        img_8x = cv2.resize(img, (int(img.shape[1]/8), int(img.shape[0]/8)), interpolation = cv2.INTER_AREA)

    #     ocr_4 = pt.image_to_string(img_4x).upper()
    #     ocr_4 = pos_proc_ocr(ocr_4)
    #     cv2.imwrite('./dataset_4/'+save_name, img_4x)
    #     ocr_6 = pt.image_to_string(img_6x).upper()
    #     ocr_6 = pos_proc_ocr(ocr_6)
    #     cv2.imwrite('./dataset_6/'+save_name, img_6x)
        ocr_8 = pt.image_to_string(img_8x).upper()
        ocr_8 = pos_proc_ocr(ocr_8)
        cv2.imwrite('./dataset_8/'+save_name, img_8x)

        original_list.append(plate_name)
        ocr_list.append(ocr_real)
    #     ocr_4_list.append(ocr_4)
    #     ocr_6_list.append(ocr_6)
        ocr_8_list.append(ocr_8)

        print(plate_name)
    except:
        pass
        #     if (ocr_real != plate_name):
#         print(ocr_real, plate_name)
```

```python
import pandas as pd
zipped_list = list(zip(original_list,ocr_list, ocr_8_list))

ocr_df = pd.DataFrame(zipped_list, columns = ['Real', 'OCR 1x', 'OCR 8x'])
ocr_df.to_csv('ocr.csv', index = False, header = True)
```
